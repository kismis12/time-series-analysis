```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank=["AXISBANK","HDFCBANK","ICICIBANK","INDUSINDBK","KOTAKBANK","SBIN"]
data=[]
for i in bank:
    data.append(pd.read_csv("stock files\\"+i+".csv"))
```

```python
data
```

```python
df=pd.concat(data)
df
```

```python
df.Symbol.value_counts() 
```

```python
#now we replace some bank names
df.replace(["KOTAKMAH","UTIBANK"],["KOTAKBANK","AXISBANK"],inplace=True)
```

```python
df.Symbol.value_counts()
```

```python
#we check the type of date 
df.Date.dtype
```

```python
# change the type of date 
df['newdate']=pd.to_datetime(df.Date)
df
```

```python
# drop the columns which we don't need or which doesn't gave some information
df.drop(['Trades','Deliverable Volume',"%Deliverble"],axis=1,inplace=True)
```

```python
axis=df.groupby('Symbol').get_group('AXISBANK')
kotak=df.groupby('Symbol').get_group('KOTAKBANK')
sbi=df.groupby('Symbol').get_group('SBIN')
icici=df.groupby('Symbol').get_group('ICICIBANK')
ind=df.groupby('Symbol').get_group('INDUSINDBK')
hdfc=df.groupby('Symbol').get_group('HDFCBANK')
```

```python
axis
```

```python
ax=axis.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
kot=kotak.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
sb=sbi.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
ici=icici.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
indus=ind.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
hd=hdfc.groupby(pd.Grouper(key='newdate',freq='Y')).mean()
```

```python
plt.figure(figsize=(18,8))
plt.plot(ax.index,ax.Open,marker='o') 
plt.plot(kot.index,kot.Open,marker='o')
plt.plot(sb.index,sb.Open,marker='o')
plt.plot(ici.index,ici.Open,marker='o')
plt.plot(indus.index,indus.Open,marker='o')
plt.plot(hd.index,hd.Open,marker='o')
plt.xlabel("year")
plt.ylabel("open")
plt.legend(['axis','kotak','sbi','icici','indus','hdfc']);
```

# #as we see there is max. hike for SBI in the year 2011 and then 2nd for HDFC ,after that opening amount for SBI goes on dec. and max. dec. from 2015 to 2016 and then goes on dec. and have min. opening amount and Kotak have max. opening amnt till no
